# With skimage.color

```
import matplotlib.pyplot as plt
import cv2
from skimage.color import rgb2hsv
# output: numpy array, shape: (height, width, 3), dtype: uint8 [0, 255]

rgb_img = cv2.imread('q10y.png')
hsv_img = rgb2hsv(rgb_img)
hue_img = hsv_img[:, :, 0]
value_img = hsv_img[:, :, 2]

fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(8, 2))

ax0.imshow(rgb_img)
ax0.set_title("RGB image")
ax0.axis('off')
ax1.imshow(hue_img, cmap='hsv')
ax1.set_title("Hue channel")
ax1.axis('off')
ax2.imshow(value_img)
ax2.set_title("Value channel")
ax2.axis('off')

fig.tight_layout()
```

#BGR2Luv

```

Luv_img = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2Luv);
plt.imshow(Luv_img)
```

```
YUV_img = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2YUV);
plt.imshow(YUV_img)
```

```

YUV_I420_img = cv2.cvtColor(rgb_img, cv2.COLOR_LBGR2Lab);
plt.imshow(YUV_I420_img)
```

# BGR2BGRA

```


BGR2BGRA_img = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2BGRA );
plt.imshow(BGR2BGRA_img)
```

# BGR2RGBA

```


BGR2RGBA_img = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2RGBA );

plt.imshow(rgb_img)
plt.figure()
plt.imshow(BGR2RGBA_img)
plt.show()
```
